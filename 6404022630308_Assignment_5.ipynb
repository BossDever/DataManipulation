```python

import pandas as pd


file_path = "/content/E0.csv"
data = pd.read_csv(file_path)


print("Step 1: Load and process the results file")
print(data.head())

```

```python

teams_df = pd.DataFrame()
teams_df['Team'] = data['HomeTeam'].unique()


teams_df['Played'] = 0
teams_df['Win'] = 0
teams_df['Lose'] = 0
teams_df['GF'] = 0
teams_df['GA'] = 0
teams_df['GD'] = 0
teams_df['Point'] = 0


print("\nStep 2: Initialize DataFrame")
print(teams_df)
```

```python

for index, row in data.iterrows():
    home_team = row['HomeTeam']
    away_team = row['AwayTeam']
    fthg = row['FTHG']
    ftag = row['FTAG']

    teams_df.loc[teams_df['Team'] == home_team, 'Played'] += 1
    teams_df.loc[teams_df['Team'] == home_team, 'GF'] += fthg
    teams_df.loc[teams_df['Team'] == home_team, 'GA'] += ftag


    teams_df.loc[teams_df['Team'] == away_team, 'Played'] += 1
    teams_df.loc[teams_df['Team'] == away_team, 'GF'] += ftag
    teams_df.loc[teams_df['Team'] == away_team, 'GA'] += fthg


    if row['FTR'] == 'H':
        teams_df.loc[teams_df['Team'] == home_team, 'Win'] += 1
        teams_df.loc[teams_df['Team'] == away_team, 'Lose'] += 1
        teams_df.loc[teams_df['Team'] == home_team, 'Point'] += 3
    elif row['FTR'] == 'A':
        teams_df.loc[teams_df['Team'] == away_team, 'Win'] += 1
        teams_df.loc[teams_df['Team'] == home_team, 'Lose'] += 1
        teams_df.loc[teams_df['Team'] == away_team, 'Point'] += 3
    else:
        teams_df.loc[teams_df['Team'] == home_team, 'Point'] += 1
        teams_df.loc[teams_df['Team'] == away_team, 'Point'] += 1

print("\nStep 3: Calculate team statistics")
print(teams_df)
```

```python

teams_df['GD'] = teams_df['GF'] - teams_df['GA']

teams_df = teams_df.sort_values(by='Point', ascending=False)


teams_df['Rank'] = range(1, len(teams_df) + 1)


print("\nStep 4: Calculate goal difference, sort, and add Rank column")
print(teams_df)
```
